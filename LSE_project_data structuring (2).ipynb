## 2. Создание файла с конкурсной массой

```python
selected_columns_massa = [
    "Номер в Книге государственной регистрации кредитных организаций",
    "URL",
    "Конкурсная масса - Общая сумма", 
    "Конкурсная масса - Динамика"
]
massa_df = merged_banks[selected_columns_massa]
massa_df.head()
```

```python
def mass_dynamic(text):
    if pd.isna(text) or text == 'Нет данных' or text == '':
        return 'Нет данных'
    pattern = r'(\d{4}):\s*([\d ]+)'
    matches = re.findall(pattern, str(text))
    result = {}
    for year, value in matches:
        clean_value = value.replace(' ', '').replace(' ', '').replace('\n', '')
        if clean_value.isdigit():
            result[year] = int(clean_value)
        else:
            result[year] = None
    return result
data = massa_df['Конкурсная масса - Динамика'].apply(mass_dynamic)
all_years = set()
for data_dict in data:
    if data_dict != 'Нет данных':
        all_years.update(data_dict.keys())
all_years = sorted(all_years, key=int)

for year in all_years:
    massa_df[f'{year}'] = data.apply(lambda x: x.get(year, np.nan) if x != 'Нет данных' and isinstance(x, dict) else 'Нет данных')
```

```python
massa_df = massa_df.drop('Конкурсная масса - Динамика', axis=1)
```

```python
massa_df.shape
```

```python
# функция для форматирования текста в числа для того, чтобы далее их было проще обрабатывать
def correct(value):
    if pd.isna(value) or value in ['Нет данных', '']:
        return np.nan
    processed = str(value).replace(' тыс. руб.', '').replace(' ', '')
    return processed if processed else np.nan
massa_df['Конкурсная масса - Общая сумма'] = massa_df['Конкурсная масса - Общая сумма'].apply(correct)
```

```python
massa_df = massa_df.fillna('Нет данных')
massa_df
```

```python
massa_df.to_excel("Конкурсная масса_динамика.xlsx", index=False)
```

## 3. Создание файла с имуществом

```python
def property(row):
    if pd.isna(row['Конкурсная масса - Имущество']) or row['Конкурсная масса - Имущество'] == 'Нет данных':
        return []
    results = []
    total = 0
    for line in str(row['Конкурсная масса - Имущество']).split('\n'):
        if ':' in line:
            cat, val = line.split(':', 1)
            clean_val = val.strip().replace(' ', '').replace(' ', '')
            value = int(clean_val) if clean_val.isdigit() else 0
            total += value
            results.append({
                'Номер в Книге государственной регистрации кредитных организаций': row['Номер в Книге государственной регистрации кредитных организаций'],
                'URL': row['URL'],
                'Имущество (активы), в том числе:': cat.strip(),
                'Балансовая стоимость на отчетную дату, тыс.руб.': value
            })
    results.append({
        'Номер в Книге государственной регистрации кредитных организаций': row['Номер в Книге государственной регистрации кредитных организаций'],
        'URL': row['URL'],
        'Имущество (активы), в том числе:': 'Итого',
        'Балансовая стоимость на отчетную дату, тыс.руб.': total
    })
    return results

property_df = []
for i, row in merged_banks.iterrows():
    property_df.extend(property(row))
property_df = pd.DataFrame(property_df)
```

## 7. Создание файла с выплатами кредиторам

```python
def creditor_payments(text, bank_id, bank_url):
    if pd.isna(text) or text in ['Нет данных', '']:
        return []
    results = []
    lines = str(text).split('\n')
    total_established = 0
    total_satisfied = 0
    total_setoff = 0 
    for line in lines:
        if ':' in line:
            queue_part, values_part = line.split(':', 1)
            queue = queue_part.strip()
            established_match = re.search(r'Установлено\s*([\d ]+)', values_part)
            satisfied_match = re.search(r'Удовлетворено\s*([\d ]+)', values_part)
            setoff_match = re.search(r'Отступное\s*([\d ]+)', values_part)
            def parse_value(match):
                if match:
                    value_str = match.group(1).replace(' ', '').replace(' ', '')
                    try:
                        return int(value_str) if value_str.isdigit() else 0
                    except:
                        return 0
                return 0
            established = parse_value(established_match)
            satisfied = parse_value(satisfied_match)
            setoff = parse_value(setoff_match)
            total_established += established
            total_satisfied += satisfied
            total_setoff += setoff
            results.append({
                'Номер в Книге государственной регистрации кредитных организаций': bank_id,
                'URL': bank_url,
                'Очередь': queue,
                'Установленно тыс. руб.': established,
                'Удовлетворенно тыс. руб.': satisfied,
                'в т.ч. отступным, тыс. руб.': setoff
            })
    results.append({
        'Номер в Книге государственной регистрации кредитных организаций': bank_id,
        'URL': bank_url,
        'Очередь': 'Итого',
        'Установленно тыс. руб.': total_established,
        'Удовлетворенно тыс. руб.': total_satisfied,
        'в т.ч. отступным, тыс. руб.': total_setoff
    })
    return results

def creditor_payments_df(input_df, payments_col, id_col, url_col):
    all_results = []
    for i, row in input_df.iterrows():
        payments_data = creditor_payments(
            row[payments_col],
            row[id_col],
            row[url_col]
        )
        all_results.extend(payments_data)
    return pd.DataFrame(all_results)
```

```python
creditors_df = creditor_payments_df(merged_banks,
    payments_col='Выплаты кредиторам - Данные',
    id_col='Номер в Книге государственной регистрации кредитных организаций',
    url_col='URL')
```

## 8. Создание файла с юридической помощью

```python
def parse_litigation_data(text, bank_id, bank_url):
    if pd.isna(text) or text in ['Нет данных', '']:
        return []
    results = []
    lines = str(text).split('\n')
    
    for line in lines:
        if '(' in line and ')' in line:
            name_part = line.split('(')[0].strip()
            data_part = line.split('(')[1].split(')')[0]
            quantity_match = re.search(r'Кол-во:\s*([\d ]+)', data_part)
            amount_match = re.search(r'Сумма:\s*([\d ]+)', data_part)
            
            def parse_numeric_value(match):
                if match:
                    value_str = match.group(1).replace(' ', '').replace(' ', '')
                    try:
                        return int(value_str) if value_str.isdigit() else 0
                    except:
                        return 0
                return 0
            
            quantity = parse_numeric_value(quantity_match)
            amount = parse_numeric_value(amount_match)
            
            results.append({
                'Номер в Книге государственной регистрации кредитных организаций': bank_id,
                'URL': bank_url,
                'Показатель': name_part,
                'Количество': quantity,
                'Сумма, тыс. руб.': amount
            })
    
    return results
```

```python
def process_litigation_df(input_df, litigation_col, id_col, url_col):
    all_results = []
    for _, row in input_df.iterrows():
        litigation_data = parse_litigation_data(
            row[litigation_col],
            row[id_col],
            row[url_col]
        )
        all_results.extend(litigation_data)
    return pd.DataFrame(all_results)

litigation_df = process_litigation_df(merged_banks,
    litigation_col='Юридическая работа - Процессуальные действия',
    id_col='Номер в Книге государственной регистрации кредитных организаций',
    url_col='URL')
```
